2개의 행렬 arr1과 arr2를 입력받아, 행렬 덧셈의 결과를 반환하는 함수, solution을 완성해주세요.

````python
def solution(a, b):
    for i in range(len(a)):
        for j in range(len(a[0])):
            a[i][j] += b[i][j]
    return a
````

### ■ [콜라츠 추측](https://programmers.co.kr/learn/courses/30/lessons/12943)
- 문제 설명  
```
1937년 Collatz란 사람에 의해 제기된 이 추측은, 
주어진 수가 1이 될때까지 다음 작업을 반복하면, 모든 수를 1로 만들 수 있다는 추측입니다.  
작업은 다음과 같습니다.
ㆍ입력된 수가 짝수라면 2로 나눕니다. 
ㆍ입력된 수가 홀수라면 3을 곱하고 1을 더합니다.
ㆍ결과로 나온 수에 같은 작업을 1이 될 때까지 반복합니다.
예를 들어, 입력된 수가 6이라면 6→3→10→5→16→8→4→2→1 이 되어 총 8번 만에 1이 됩니다.
위 작업을 몇 번이나 반복해야하는지 반환하는 함수, solution을 완성해 주세요.
단, 작업을 500번을 반복해도 1이 되지 않는다면 –1을 반환해 주세요.
```
- 제한 조건  
```
ㆍ입력된 수, num은 1 이상 8000000 미만인 정수입니다.
```
- 예시

---
입출력 예시  
image.png

---

````python
def solution(num):
    answer = 0
    while num > 1:
        if num %2 == 0:
            num //= 2
        else:
            num = num*3 +1
        answer += 1
    if answer >= 500:
        return -1
    else:
        return answer
````

### ■ [최대공약수와 최소공배수](https://programmers.co.kr/learn/courses/30/lessons/12940)
- 문제 설명  
```
두 수를 입력받아 두 수의 최대공약수와 최소공배수를 반환하는 함수, solution을 완성해 보세요. 
배열의 맨 앞에 최대공약수, 그다음 최소공배수를 넣어 반환하면 됩니다. 
예를 들어 두 수 3, 12의 최대공약수는 3, 최소공배수는 12이므로 solution(3, 12)는 [3, 12]를 반환해야 합니다.
```
- 제한 조건  
```
ㆍ두 수는 1이상 1000000이하의 자연수입니다.
```
- 예시

---
입출력 예시  
image.png

---

````python
def solution(n, m):
    cd = []
    for i in range(1,max(n,m)):
        if n%i == 0 and m% i == 0 :
            cd.append(i)
        else:
            pass
    gcd = max(cd)
    lcm = n*m//gcd
    return [gcd, lcm]
````

def gcdlcm(a, b):
    c, d = max(a, b), min(a, b)

    while t > 0:
        t = c % d
        c, d = d, t
    answer = [c, int(a*b/c)]

    ㆍmath.floor(x) : x를 내림하여 정수값을 반환한다.
    ㆍmath.fabs(x) : x의 절댓값을 반환한다.
    ㆍmath.trunc(x) : x의 소수점을 자른 후 정수값을 반환합니다.
    ㆍmath.copysign(x, y) : x값의 절댓값에 y의 부호를 갖는 유리수를 반환한다.
    ㆍmath.comb(n, k) : nCk과 같은 조합 값을 반환한다. (n개의 수에서 k개를 선택)
    ㆍmath.perm(n, k) : nPk와 같은 순열 값을 반환한다. (n개의 수에서 k를 꺼내 순서대로 나열)
    ㆍmath.factorial(x) : x의 팩토리얼을 반환합니다. (1*2*3*...*x)
    ㆍmath.fmod(x, y) : x에 y를 나눈 후 나머지를 제공한다. (부동소수점 연산에 사용)
    ㆍmath.frexp(x) : x를 (m * 2 ** e) 형태로 반환합니다.
    ㆍmath.ldexp(x, i) : x * 2 ** i 값을 반환합니다.
    ㆍmath.fsum(iterable) : 이터러블(iterable)에 있는 값의 정확한 부동 소수점 합을 반환합니다.
    ㆍmath.prod(iterable) : 이터러블(iterable)의 모든 수의 곱을 반환합니다. (빈 경우 1을 반환)
    ㆍmath.gcd(integers) : 여러개의 정수를 받아 최대 공약수를 반환합니다.
    ㆍmath.lcm(integers) : 여러개의 정수를 받아 최소 공약수를 반환합니다.
    ㆍmath.isclose(a, b) : a의 실수 연산이 b와 같은지 확인하고 bool 값을 반환합니다.
    ㆍmath.isfinite(x) : x값이 무한하거나 Nan(Not a number) 일경우 True 값을 반환합니다.
    ㆍmath.isinf(x) : x가 양 또는 음의 무한대이면 True, 반대일 경우 False를 반환합니다.
    ㆍmath.isnan(x) : x가 Nan(Not a number)일경우 True, 반대일 경우 False를 반환합니다.
    ㆍmath.isqrt(n) : n의 음이 아닌 정수 제곱근을 반환합니다.
    ㆍmath.modf(x) : x의 (소수, 정수)부분으로 반환합니다. 정수부분은 float로 반환됩니다.
2) 지수와 로그함수
    ㆍmath.exp(x) : 자연상수인 e의 x 거듭제곱을 반환합니다. → 반환값 :  e^x
    ㆍmath.expm1(x) : 자연상수인 e의 x 거듭제곱에서 1을 뺀값을 반환합니다. → 반환값 :  e^x - 1
    ㆍmath.log(x) : 진수가 x인 자연로그를 반환합니다. → 반환값 :  ln(x)
    ㆍmath.log(x,base) : 밑을 base, 진수를 x로 가지는 log를 반환합니다. → 반환값 :  log(x) / log(base)
    ㆍmath.log1p(x) : 밑이 e(자연상수)인 x+1의 자연로그를 반환합니다. → 반환값 :  ln(1+x)
    ㆍmath.log2(x) : 밑이 2인 로그를 반환합니다. → 반환값 :  log(x) / log(2)
    ㆍmath.log10(x) : 밑이 10인 상용로그를 반환합니다. → 반환값 :  log(x) / log(10)
    ㆍmath.pow(x, y) : x의 y 거듭제곱을 반환합니다. → 반환값 :  x^y
    ㆍmath.sqrt(x) : x의 제곱근을 반환합니다. → 반환값 :  x^(1/2)
